```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
url = "https://raw.githubusercontent.com/rahulinchal/SPPU/refs/heads/main/Data/data_YesBank_StockPrices.csv"
df = pd.read_csv(url)
df
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['Open', 'High', 'Low', 'Close']].corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Feature Correlations')
plt.show()
```

```python
df[['Open', 'High', 'Low', 'Close']].hist(bins=20, figsize=(10,8))
plt.suptitle('Histograms of Features')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Open'], df['Close'], color='blue')
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.title('Scatter Plot: Open vs Close')
plt.show()
```

```python
df.dropna(inplace=True)
df['Date'] = pd.to_datetime(df['Date'], format='%b-%y', errors='coerce')
df.dropna(inplace=True)
df['Month'] = df['Date'].dt.month
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['Open', 'High', 'Low', 'Close', 'Month']].corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Feature Correlations')
plt.show()
```

```python
df[['Open', 'High', 'Low', 'Close']].hist(bins=20, figsize=(10,8))
plt.suptitle('Histograms of Features')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Open'], df['Close'], color='blue')
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.title('Scatter Plot: Open vs Close')
plt.show()
```

```python
X = df[['Open', 'High', 'Low', 'Month']]
y = df['Close']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)
r2_lr = r2_score(y_test, y_pred_lr)
print("Linear Regression R2 Score:", r2_lr)
print("Linear Regression Accuracy:", r2_lr * 100, "%")
```

```python
knn_model = KNeighborsRegressor()
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
r2_knn = r2_score(y_test, y_pred_knn)
print("\nKNN Regressor R2 Score:", r2_knn)
print("KNN Regressor Accuracy:", r2_knn * 100, "%")
```

```python
svr_model = SVR()
svr_model.fit(X_train, y_train)
y_pred_svr = svr_model.predict(X_test)
r2_svr = r2_score(y_test, y_pred_svr)
print("\nSVR R2 Score:", r2_svr)
print("SVR Accuracy:", r2_svr * 100, "%")
```

```python
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
r2_rf = r2_score(y_test, y_pred_rf)
print("\nRandom Forest R2 Score:", r2_rf)
print("Random Forest Accuracy:", r2_rf * 100, "%")
```
